```python
try:
    from qiskit import QuantumCircuit
except Exception as e:
    print('Qiskit import failed:', e)
    print('This notebook expects a clean Qiskit environment. Create a new virtual environment and install a compatible Qiskit version (see https://qisk.it/packaging-1-0).')
else:
    qc = QuantumCircuit(3)
    qc.h(0); qc.cx(0,1); qc.cx(0,2)
    print(qc)            # always works (text)
    # try mpl (needs matplotlib + pylatexenc installed)
    qc.draw('mpl')
```
